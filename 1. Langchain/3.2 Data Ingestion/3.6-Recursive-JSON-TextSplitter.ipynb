```python
import json 
import requests

json_data = requests.get('https://api.smith.langchain.com/openapi.json').json()
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=300)
json_text_split = json_splitter.split_json(json_data)
json_text_split
```

```python
for i in json_text_split[:2]:
    print(i)
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=300)
json_text_split = json_splitter.create_documents([json_data])
json_text_split
```

```python
for item in json_text_split[:3]:
    print(item)
```
