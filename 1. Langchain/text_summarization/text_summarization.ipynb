```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain


load_dotenv()

groq_api_key = os.getenv('GROQ_API_KEY')
llm = ChatGroq(
    model="Gemma2-9b-It",
    groq_api_key=groq_api_key
)
llm

```

```python
from langchain.schema import AIMessage, HumanMessage, SystemMessage

messages = [
    SystemMessage(content="you are expert with expertize of summarizing speech."),
    HumanMessage(content="please provide a short and consise summary of the following text: {text}")
]

llm.get_num_tokens(text)
```

```python
generic_template = """write a summary of the following speech: {text}

Translate the pricise summary into {language}"""

prompt = PromptTemplate(
    input_variables=['text', 'language'],
    template=generic_template
)

complete_prompt = prompt.format(text=text, language='hindi')

llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
response = llm_chain.run(
    {
        "text": text,
        "language": "hindi"
    }
)
response
```

```python
# stuff document chain text summarization

from langchain.chains.summarize import load_summarize_chain
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader(r"D:\ved_glory\Langchain\Langchain\1. Langchain\text_summarization\apjspeech.pdf")
documents = loader.load_and_split()

template = "write a consice short summary of the following text: {text}"
prompt = PromptTemplate(input_variables=["text"], template=template)

chain = load_summarize_chain(llm=llm, prompt=prompt,chain_type="stuff", verbose=True)

response = chain.run(documents)
response
```

```python
# map reduce to summarize large documents

from langchain_text_splitters import RecursiveCharacterTextSplitter

splitted_docs = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200).split_documents(documents=documents)

chunk_prompt = """please summarize the below text : {text} \n\n summary:"""

mapped_chunk_prompt = PromptTemplate(
    input_variables=['text'],
    template=chunk_prompt
)

summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt = mapped_chunk_prompt
)

final_prompt = """please provide the final summary of the entire speech with these important ponits and motivation title, start the precise summary with an introduction and provide the summary number points for the speech : {text}"""

final_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=final_prompt
)

summary_chain_map_reduce = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt = mapped_chunk_prompt,
    combine_prompt = final_prompt_template,
    verbose=True
)

output_summary = summary_chain_map_reduce.run(splitted_docs)
output_summary
```
