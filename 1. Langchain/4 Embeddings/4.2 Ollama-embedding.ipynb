##### https://github.com/ollama/ollama

```python
from langchain_community.embeddings import OllamaEmbeddings

ollama_gemma2_embeddings = OllamaEmbeddings(model='gemma2:2b')

gemma2_embedding_results = ollama_gemma2_embeddings.embed_documents(
        [
            "Alpha is the first letter of Greek alphabet",
            "Beta is the second letter of Greek alphabet",   
        ]
)

gemma2_embedding_results

```

```python
gemma2_embedding_results[1]
```

```python
ollama_gemma2_embeddings.embed_query("what is the second letter of Greek alphabet")
```

```python
len(gemma2_embedding_results[1])
```
