### Embeddings

Convert Text into vector is called embeddings.

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
SAMPLE_TEXT = "This is sample text for checking openai embeddings."

query_result = embeddings.embed_query(text=SAMPLE_TEXT)
query_result
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings

embeddings_1024 = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=1024)
embeddings_1024
```

```python
query_result_1024 = embeddings_1024.embed_query(text=SAMPLE_TEXT)
query_result_1024
```

```python
len(query_result_1024)
```

### Load text file and create documnets

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
# splitting Documnets

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
final_docs = text_splitter.split_documents(docs)
final_docs
```

### Save embedding into Vectordb (Cromadb)

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_docs, embeddings_1024)
db
```

```python
search_query = "these issues have previously been explored by myth"

retrieved_results = db.similarity_search(query=search_query)
print(retrieved_results)
```
